A vector is just a list of numbers — like [0.2, -0.8, 0.5, 1.3, ...]. In AI, these numbers are carefully learned so that similar things end up with similar vectors.

Think of it like GPS coordinates, but for meaning. "Dog" and "puppy" will have very close coordinates, while "dog" and "airplane" will be far apart.

```python
# A simple 3D vector for the word "king"

king  = [0.9,  0.1,  0.8]
queen = [0.8,  0.9,  0.8]
man   = [0.9,  0.1,  0.1]
woman = [0.8,  0.9,  0.1]

# Famous relationship:
# king - man + woman ≈ queen
```
